```python
import numpy as np
from numpy.random import default_rng
```

```python
rng = default_rng(93922) # Generate random number generator
```

```python
import math
random_normal_men_heights = rng.normal(174, 5.5, size=1000) # Generate random avg height
random_normal_men_weights = rng.normal(75,9, size =1000) # Generate random avg weight

heights = random_normal_men_heights
weights = random_normal_men_weights

heights.sort()
weights.sort()

print(heights)
print(weights)
```

```python
def my_percentile():
    #  my_h should be in cm.
    #  my_w  should be in kg.
    my_h = float(input("Input your height in cm : "))
    my_w = float(input("Input your weight in kg : "))
    while True:
        if (isinstance(my_h, float) == False or my_h < 0): 
            my_h = float(input("Need valid height in centimeters! : "))
        if (isinstance(my_w, float)== False or my_w < 0):
            my_w = float(input("Need Valid weight in kilograms! : "))

        else:
            break



    percentile_msg = "Quick Note, an i th percentile means that out of 100 random people, you are the i th person in comparison to other individuals"
    print(percentile_msg)
    print("\n")

    my_h_percentile = 0
    my_w_percentile = 0

    height_acc = 0
    weight_acc = 0

    for i in range(len(heights)):
        if heights[i] >= my_h:
            my_h_percentile = math.ceil(((len(heights) - height_acc) / len(heights)) * 100)
            if my_h_percentile == 1:
                print("Your height " + str(my_h) + "cm stands at the " + str(my_h_percentile) + "st percentile")
                break
            elif my_h_percentile == 2:
                print("Your height " + str(my_h) + "cm stands at the " + str(my_h_percentile) + "nd percentile")
                break
            elif my_h_percentile == 3:
                print("Your height " + str(my_h) + "cm stands at the " + str(my_h_percentile) + "rd percentile")
                break
            else:
                print("Your height " + str(my_h) + "cm stands at the " + str(my_h_percentile) + "th percentile")
                break
            
        else:
            height_acc += 1  
    
    for j in range(len(weights)):
        if weights[j] >= my_w:
            my_w_percentile = math.ceil(((len(weights) - weight_acc) / len(weights)) * 100)

            if my_w_percentile == 1:
                print("and your weight " + str(my_w) +"kg at the " + str(my_w_percentile) + "st percentile.")
                break
            elif my_w_percentile == 2:
                print("and your weight " + str(my_w) +"kg at the " + str(my_w_percentile) + "nd percentile.")
                break
            elif my_w_percentile == 3:
                print("and your weight " + str(my_w) +"kg at the " + str(my_w_percentile) + "rd percentile.")
                break
            else:
                
                print("and your weight " + str(my_w) +"kg at the " + str(my_w_percentile) + "th percentile")
                print("out of 1000 random sample of an adult men.")

                avg_msg = "Considering that the average is " + str(math.floor(np.mean(heights))) + "cm."

                if my_h_percentile <= 15:
                    print("You are considered tall.")
                    print(avg_msg)
                elif (my_h_percentile > 15) and (my_h_percentile <= 40):
                    print("You are considered slightly taller than the average height.")
                    print(avg_msg)
                elif (my_h_percentile > 40) and (my_h_percentile < 60):
                    print("You are considered to be an average height")
                    print(avg_msg)
                elif (my_h_percentile >= 60) and (my_h_percentile < 85):
                    print("You are considered slightly shorter than average")
                    print(avg_msg)
                else:
                    print("You are considered short")
                    print(avg_msg)
                break

        else:
            weight_acc += 1



# Edge cases that need to be dealt with: string input

# call my_percentile() to use the function.
my_percentile()


```

```python
import math

# 5 point summary data


def height_data_summary(list):
    print("The five number summary of this data is\n")
    print("Min : " + str(math.ceil((list[0])*100)/100) + "cm\n")
    print("First Quantile : " + str(math.ceil(np.median(list[:len(list)//2])*100)/100) + "cm\n")
    print("Meidan : " + str(math.ceil((np.median(list))*100)/100) + "cm\n")
    print("Third Quantile : " + str(math.ceil(((np.median(list[len(list)//2:]))*100)/100)) + "cm\n")
    print("Max :" + str(math.ceil((list[len(list)-1])*100)/100) + "cm\n")


def weight_data_summary(list):
    print("The five number summary of this data is\n")
    print("Min : " + str(math.ceil((list[0])*100)/100) + "kg\n")
    print("First Quantile : " + str(math.ceil(np.median(list[:len(list)//2])*100)/100) + "kg\n")
    print("Median : " + str(math.ceil((np.median(list))*100)/100) + "kg\n")
    print("Third Quantile : " + str(math.ceil(((np.median(list[len(list)//2:]))*100)/100)) + "kg\n")
    print("Max :" + str(math.ceil((list[len(list)-1])*100)/100) + "kg\n")


height_data_summary(heights)
print("-----------------------------------------------------")
weight_data_summary(weights)
```
